### 1.加载包和数据集

```python

import tensorflow as tf  #  1.11.0 版本
import numpy as np
import pandas as pd
import math
from sklearn.metrics import accuracy_score,confusion_matrix
from src.utils.dnn_framework import DNN
from src.utils.cnn_framework import CNN
from  src.utils.parse_mnist import  parse_mnist_data
import copy
```

### 2.加载数据和数据处理

```python
train_x,train_label,test_x,test_label = parse_mnist_data()

train_X = train_x/255
train_y = DNN.one_hot(Y = np.squeeze(train_label),depth=10,axis=1)

test_X = test_x/255
test_y = DNN.one_hot(Y = np.squeeze(test_label),depth=10,axis=1)

print('训练集X:',str(train_X.shape))
print('训练集Y:',str(train_y.shape))
print('测试集X:',str(test_X.shape))
print('测试集Y:',str(test_y.shape))
```

### 3.DNN网络建模
#### 分别用'gd','momentum','adam'优化算法训练模型

```python

DNN_model = DNN(iters = 1000,
                batch_size = 128,
                task_type = 'multi',
                learning_rate = 0.0001,
                regularizer_type = 'l2',
                regularizer_rate = 0.002,
                optimizer = 'adam',
                layer_dims = [100,20,10],
                active_functions = ['relu','relu','softmax']
               )

DNN_model
```

```python
optimizers = ['gd','momentum','adam']

model_info = {}
for optimizer in optimizers:
    model_info_tmp = {}
    DNN_model.set_params(**{'optimizer':optimizer})
    DNN_model.fit(data = {'X':train_X.T,'Y':train_y.T})
    
    train_label_pre = DNN_model.predict(X = train_X.T)
    accuracy = accuracy_score(y_true = np.squeeze(train_label),y_pred = train_label_pre)
    model_info_tmp.update({'train_acc':accuracy})
    
    test_label_pre = DNN_model.predict(X = test_X.T)
    accuracy = accuracy_score(y_true = np.squeeze(test_label),y_pred = test_label_pre)
    model_info_tmp.update({'test_acc':accuracy})
    
    model_info_tmp.update({'model_conf':DNN_model.__dict__})
    model_info[optimizer] = model_info_tmp
    
```

#### 从测试集准确度来看，优化结算法排序为：adam>momentum>gd（梯度下降）

optimizer|train_acc|test_acc
:-|:-:|--
gd|0.9263|0.9293
momentum|0.9798|0.9745
adam|0.9893|0.9794

```python

for k,v in model_info.items():
    outstr = '优化算法:{k}，train_acc:{train_acc},test_acc:{test_acc}'.\
    format(k = k,train_acc = round(v['train_acc'],4),test_acc = round(v['test_acc'],4))
    print(outstr)
```

####  网络结构改为[100,50,10]，进行学习率衰减训练，记录各迭代次数的模型和模型效果

```python

DNN_model = DNN(iters = 1000,
                batch_size = 128,
                task_type = 'multi',
                learning_rate = 0.0001,
                regularizer_type = 'l2',
                regularizer_rate = 0.002,
                optimizer = 'adam',
                learning_decay = True,
                learning_decay_rate = 0.9,
                layer_dims = [100,50,10],
                active_functions = ['relu','relu','softmax']
               )

DNN_model
```

```python

import math
learning_base = 0.0001
decay_rate = 0.9
train_acc = {}
test_acc = {}
DNN_models = []

for i in range(1000):
    decay_step = int(i/50)
    learning_rate = learning_base * math.pow(decay_rate,decay_step)
    DNN_model.set_params(learning_rate = learning_rate,iters = 1,layer_dims = [100,50,10],active_functions = ['relu','relu','softmax'])
    print('iters:',i)
    
    DNN_model.fit(data={'X':train_X.T,'Y':train_y.T},init_param_seed = 1)
    
    print('compute acc')
    train_label_pre = DNN_model.predict(X = train_X.T)
    accuracy = accuracy_score(y_true = np.squeeze(train_label),y_pred = train_label_pre,normalize = True)
    accuracy = np.round(accuracy,4)
    train_acc[i] = accuracy
    
    test_label_pre = DNN_model.predict(X = test_X.T)
    accuracy = accuracy_score(y_true = np.squeeze(test_label),y_pred = test_label_pre,normalize = True)
    accuracy = np.round(accuracy,4)
    test_acc[i] = accuracy
    DNN_models.append(copy.deepcopy(DNN_model))
    
```

#### 保存最优模型

```python

acc_lst = list(test_acc.values())
idx = acc_lst.index(max(acc_lst))

print('训练集acc:',train_acc[idx])
print('测试集acc:',test_acc[idx])

DNN.save_model(model = DNN_models[idx],filename = 'model/mnist-model-dnn.pkl')
```

#### 加载模型，计算模型效果和混淆矩阵
模型加载和效果计算

```python

DNN_model = DNN.load_model(filename = 'model/mnist-model-dnn.pkl')

train_label_pre = DNN_model.predict(X=train_X.T)
accuracy = accuracy_score(y_true = np.squeeze(train_label),y_pred=train_label_pre,normalize = True)
accuracy = np.round(accuracy,4)
print('训练集的准确度:' + str(accuracy))

test_label_pre = DNN_model.predict(X = test_X.T)
accuracy = accuracy_score(y_true = np.squeeze(test_label),y_pred = test_label_pre,normalize = True)
accuracy = np.round(accuracy,4)
print('测试集的准确度:' + str(accuracy))
```

训练集混淆矩阵

```python

train_confusion_matrix = confusion_matrix(y_true = np.squeeze(train_label),y_pred = train_label_pre)
acc_rate = np.diag(train_confusion_matrix)/train_confusion_matrix.sum(axis = 1)
train_confusion_matrix = pd.DataFrame(train_confusion_matrix)
train_confusion_matrix['acc_rate'] = acc_rate
train_confusion_matrix
```

测试集混淆矩阵

```python

test_confusion_matrix = confusion_matrix(y_true = np.squeeze(test_label),y_pred = test_label_pre)
acc_rate = np.diag(test_confusion_matrix)/test_confusion_matrix.sum(axis = 1)
test_confusion_matrix = pd.DataFrame(test_confusion_matrix)
test_confusion_matrix['acc_rate'] = acc_rate
test_confusion_matrix
```
